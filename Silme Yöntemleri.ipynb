# SİLME YÖNTEMLERİ

```python
import numpy as np
import pandas as pd
V1=np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2=np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3=np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df=pd.DataFrame({
    "V1":V1,
    "V2":V2,
    "V3":V3
})
df
```

```python
#en az 1 tane boş değişken varsa sil
df.dropna()
```

```python
#tüm gözlem değerleri NaN olan değerlerin silinmesi
df.dropna(how="all")
```

```python
#değişkenler bazında silmek istersek.
df.dropna(axis=1)
```

```python
#tüm değerleri NaN olan değişkenleri(sütünları) silmek için ise; ama hiçbirinde öyle argüman yoktu örnek olsun.maksat öğrenmek
df.dropna(axis=1,how="all")
```

```python
#hadi bir örnek olsun
df["sil_beni"]=np.nan
```

```python
#sil beni boş olduğu için o silindi.
df.dropna(axis=1,how="all",inplace=True)
```
